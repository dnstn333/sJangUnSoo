```python
import re
regularExpression = re.compile('[^a-zA-Z]*')

twit = []
classVec = []
DocList = []

file = open('Twit.txt')
for i in file.readlines():
    twit.append(i)
idx = len(twit)

for i in range(idx):
    temp = regularExpression.split(twit[i])
    DocList.append(temp)

file =open('ClassVec.txt')
for i in file.readlines():
    classVec.append(i.rstrip('\n'))
    
    
print classVec
print DocList
```

```python
ExcepWordList = ["Darvish","Osaka","YuDarvish","Rangers","mets","Mark","Trumbo","DanielMurphy","Daniel","Murphy",
                 "seunghwanoh",'seung', 'hwan',"cardinals","stlouis","manny","machado", "orioles","mlb",
                "park","twins","maeda","kenta","dodgers","la","ladodgers","debut","hyunsookim",
                "kim","hyun","soo","baltimore","korean","tyler","colins","tigers", "Orioles",
                 "tylercollins","keuchel","houstonastros","giants","jepsen","braves",
                "cubs","redSox","greinke","puig","molina","detroitTigers","a","an","the","we","it","his","him",
                "for","by","and","of","or","to","is","are","has","have","about","i","you",
                "he","they","MLB","baseball",'']

for i in range(idx):
    text = []
    for word in DocList[i]:
        word = word.upper().lower()
        if word not in ExcepWordList:
            text.append(word)
   
    DocList[i] = text

print DocList
```

```python
%cd ../../src/ch04
```

```python
import bayes

myVocabList = bayes.createVocabList(DocList)

trainMat=[]
int_classVec = []

for postinDoc in DocList:
    trainMat.append(bayes.setOfWords2Vec(myVocabList, postinDoc))
    
for str_class in classVec:
    int_class = int(str_class)
    int_classVec.append(int_class)

p0V,p1V,pAb=bayes.trainNB0(trainMat,int_classVec)
print p0V
print p1V
print pAb
```

## 1. 나이브 베이즈 분류기 평가

```python
import numpy

regularExpression = re.compile('[^a-zA-Z]*')

file =open('validation_Twit.txt')
testDoc=[]
for i in file.readlines():
    testDoc.append(i)
    
file =open('validation_ClassVec.txt')
classList=[]
for i in file.readlines():
    classList.append(i)

errorCount = 0
docIdx = 0;
for doc in testDoc:
    testEntry = regularExpression.split(doc)
    thisDoc = numpy.array(bayes.setOfWords2Vec(myVocabList, testEntry))
    classifyed = bayes.classifyNB(thisDoc,p0V,p1V,pAb)
    if classifyed != int(classList[docIdx]):
        errorCount += 1
    print 'Gass is ', classifyed, 'and Real is ', classList[docIdx]
    docIdx += 1
    
print 'the error rate is: ',float(errorCount)/len(testDoc)
```

## 2. SVM 분류기 평가

```python
import svmMLiA
```

```python
classSVM = []

#class 값 0 or 1  ->  -1 or 1 로 변환
for strVec in classVec:
    intVec = int(strVec)
    if intVec == 1:
        classSVM.append(int(1))
    else :
        classSVM.append(int(-1))

b, alphas = svmMLiA.smoP(trainMat, classSVM, 200, 0.001, 1000)
print b, alphas
```

```python
ws = svmMLiA.calcWs(alphas, trainMat, classSVM)
trainDataMat = numpy.mat(trainMat)
classSVMMat = numpy.mat(classSVM).transpose()

indexSV = numpy.nonzero(alphas.A > 0)[0]
SVs = trainDataMat[indexSV]
labelSV = classSVMMat[indexSV]

print "number of Support Vectors : %d" %numpy.shape(SVs)[0] 
```

```python
cntError = 0
m = numpy.shape(trainDataMat)[0]
for i in range(m):
    predict = trainDataMat[i] * numpy.mat(ws) + b
    if numpy.sign(predict) != numpy.sign(int(classSVMMat[i])):
        cntError += 1
    print 'Gass is', numpy.sign(predict), 'Real is ', classSVM[i]
print 'Total error rate : ', float(cntError) / m
```

```python
#위와 동일한 진행을 testSet에 진행하여 인식률 확인
testSVM = []
for doc in testDoc:
    arr = numpy.array(bayes.setOfWords2Vec(myVocabList, doc))
    testSVM.append(arr)
testSVMMat = numpy.mat(testSVM)

for i in range(len(classList)):
    if classList[i] == 0:
        classList[i] = '-1'

errorCount = 0
m = numpy.shape(testSVMMat)[0]
for i in range(m):
    predict = trainDataMat[i] * numpy.mat(ws) + b
    if numpy.sign(predict) != numpy.sign(int(classList[i])):
        errorCount += 1
    print 'Gass is ', numpy.sign(predict), 'and Real is ', classList[i]
    
print 'the error rate is: ',float(errorCount)/m
```
